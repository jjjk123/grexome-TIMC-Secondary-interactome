```python
# load scores
scores = {}

for i in range(3, 11):
    tmp = {}
    with open(f"scores_d{i}.tsv", 'r') as f:
        data = f.read().splitlines()
        for line in data:
            line_splitted = line.split('\t')
            gene, score = line_splitted
            tmp[gene] = score
    
    scores[i] = tmp
```
